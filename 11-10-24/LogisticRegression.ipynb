```python
df=pd.read_csv("/content/drive/MyDrive/nmims/sem1/aiml/kshamta mam/11-10-24/heart.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
y_pred1=clf1.fit(X_train,y_train).predict(X_test)
y_pred2=clf2.fit(X_train,y_train).predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print("Accuracy of Logistic Regression:",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Tree:",accuracy_score(y_test,y_pred2))
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
confusion_matrix(y_test,y_pred2)
```

```python
from sklearn.metrics import recall_score, precision_score,f1_score
```

```python
print("Precission: ",precision_score(y_test,y_pred1))
print("Recall: ",recall_score(y_test,y_pred1))
print("F1 Score: ",f1_score(y_test,y_pred1))
```
