```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from mlxtend.frequent_patterns import apriori, association_rules

```

```python
df=pd.read_csv("/content/drive/MyDrive/nmims/sem1/aiml/kshamta mam/22-11-24/bread basket.csv")
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
g_df=df.groupby(['Transaction','Item'])['Item'].count().reset_index(name='Count')
```

```python
basket_df=g_df.pivot_table(index='Transaction',columns='Item',values='Count',aggfunc='sum').fillna(0)
```

```python
g_df.head()
```

```python
basket_df.head()
```

```python
#one hot encoding by defining a function to make data suitable for concerned libraries
#not for the purpose of categorical to numerical this encoding is for differnt purpose
def hot_encode(x):
  if x<=0:
    return 0
  if x>=1:
    return 1
```

```python
basket_encoded=basket_df.applymap(hot_encode)
basket_df=basket_encoded
```

```python
basket_df.head()
```

```python
#data mining

frq_items=apriori(basket_df,min_support=0.01,use_colnames=True)
```

```python
frq_items
```

```python

rules=association_rules(frq_items,metric='confidence',min_threshold=0.25)
```

```python
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
```
